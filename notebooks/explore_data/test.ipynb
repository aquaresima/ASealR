```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
project_folder = "/home/alequa/Documents/Research/phd_project/speech/ASealR/"

data = pd.read_csv(os.path.join(project_folder, "data/vocalizations","vocalizations_info.csv"))
data.Filepath[1]
```

```python
def get_spikes(audio_file, encoding=350):
    p= audio_file.split('.')[-3:][0] + '_spike_train'+str(encoding)+'.npz'
    p = os.path.join(project_folder,  p)
    assert(os.path.exists(p))
    return p
```

```python
p =get_spikes(data.Filepath[0])

call = np.load(p)['arr_0']
units = call[1]
times = call[0]
 
 
# A quick raster plot for one of the samples
 
fig = plt.figure(figsize=(16,4))
fig, ax = plt.subplots()
ax.scatter(times,units, color="k", alpha=0.33, s=2)
```
